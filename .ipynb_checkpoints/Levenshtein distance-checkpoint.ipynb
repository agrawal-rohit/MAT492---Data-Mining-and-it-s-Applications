```python
# Rohit Agrawal
# 1610110292
# MAT492 - Data Mining and its Applications

import numpy as np

word1 = "intention"
word2 = "execution"
word1length = len(word1)
word2length = len(word2)

# Add '#' to each word to form the Edit matrix
word1list = list(word1)[::-1]
word1list.append('#')
word1list = word1list[::-1]
word2list = list(word2)[::-1]
word2list.append('#')
word2list = word2list[::-1]

EDmatrix = np.zeros((word1length + 1, word2length + 1))
EDmatrix[:,0] = np.arange(word1length+1)[::-1]
EDmatrix[-1,:] = np.arange(word2length+1)
print('Edit Distance Matrix = \n' + str(EDmatrix))
```

```python
# Compute all distances for each entry in the EDmatrix and fill it with the one thats the minimum 
for i in range(0,EDmatrix.shape[0]-1)[::-1]:
    for j in range(1, EDmatrix.shape[1]):
        print('i = ' + str(i) + ', j = ' + str(j))
        dist1 = EDmatrix[i+1,j] + 1
        dist2 = EDmatrix[i,j-1] + 1
        dist3 = EDmatrix[i+1,j-1]
        print('Comparing characters ' + str(word1list[EDmatrix.shape[0] - i - 1]) + ' and ' + str(word2list[j]))
        if word1list[EDmatrix.shape[0] - i - 1] == word2list[j]:
            dist3 += 0
        else:
            dist3 += 2
        print('Distance 1 = ' + str(dist1) + '\n')
        print('Distance 2 = ' + str(dist2) + '\n')
        print('Distance 3 = ' + str(dist3) + '\n')
        EDmatrix[i,j] = min([dist1, dist2, dist3])
```

```python
# Computed Edit Distance Matrix
print('Edit Distance Matrix = \n' + str(EDmatrix) + '\n')
print('-------------------------------------\n')

# Number of operations 
print("Optimal number of operations needed = " + str(EDmatrix[0 , EDmatrix.shape[1] - 1]) + '\n')
print('-------------------------------------\n')

# Optimal path
print('Optimal path -> \n')
i = 0
j = EDmatrix.shape[1] - 1
while(j>=0 and i<EDmatrix.shape[0]):
    print('(' + str(i) + ',' + str(j) + ')')
    curr_val = EDmatrix[i, j] 
    dist1 = EDmatrix[i, j-1] 
    if(i == EDmatrix.shape[0] - 1):      # If path is at the edge of the matrix, we simply move to the only block available to us
        j = j-1
        continue
    else:
        dist2 = EDmatrix[i+1, j]
        dist3 = EDmatrix[i+1, j-1]
    
    idx = np.argmin([dist1, dist2, dist3])          # Find which distance is the minimum
    if idx == 0:
        j = j-1
    if idx == 1:
        i = i+1
    if idx == 2:
        i = i+1
        j = j-1
print('\n-------------------------------------\n')

# Step by step operations
testword = word1list
i = 0
j = EDmatrix.shape[1] - 1
n = EDmatrix.shape[0] - 1

counter = 1
print('Step by step operations -> \n')
while(j>=0 and i<EDmatrix.shape[0]):
    curr_val = EDmatrix[i, j] 
    dist1 = EDmatrix[i, j-1] 
    if(i == EDmatrix.shape[0] - 1):
        j = j-1
        if(EDmatrix[i][j] == curr_val):
            #print(str(counter) + '. No Change \n')
            #counter += 1
            continue
        else:
            if word2list[j+1] is not '#':
                print(str(counter) + '. Insert ' + str(word2list[j+1]) + '\n')
                counter += 1
        continue
    else:
        dist2 = EDmatrix[i+1, j]
        dist3 = EDmatrix[i+1, j-1]
    
    idx = np.argmin([dist1, dist2, dist3])
    if idx == 0:
        j = j-1
        if(EDmatrix[i][j] == curr_val):
            #print(str(counter) + '. No Change \n')
            #counter += 1
            continue
        else:
            print(str(counter) + '. Insert ' + str(word2list[j+1]) + '\n')
            counter += 1
    if idx == 1:
        i = i+1
        if(EDmatrix[i][j] == curr_val):
            #print(str(counter) + '. No Change \n')
            #counter += 1
            continue
        else:
            print(str(counter) + '. Delete ' + str(testword[n-i+1]) + '\n')
            counter += 1
    if idx == 2:
        i = i+1
        j = j-1
        if(EDmatrix[i][j] == curr_val):
            #print(str(counter) + '. No Change \n')
            #counter += 1
            continue
        else:
            print(str(counter) + '. Substitute ' + str(testword[n-i+1]) + ' with ' + str(word2list[j+1]) + ' \n')
            counter += 1

print('-------------------------------------\n')
print('Original string - ' + str(word1) + '\n')
print('Final string - ' + str(word2) + '\n')
```
