```python
# Lp norm
def lp_norm(X,Y,p):
    return np.power((np.abs(X)**p + np.abs(Y)**p), 1/p)
```

```python
v1 = np.linspace(-1,1,1000)
v2 = np.linspace(-1,1,1000)

# L1 norm
X1,Y1 = np.meshgrid(v1, v2)
Z1 = lp_norm(X1,Y1,1)

# L2 norm
X2,Y2 = np.meshgrid(v1, v2)
Z2 = lp_norm(X2,Y2,2)

# L3 norm
X3,Y3 = np.meshgrid(v1, v2)
Z3 = lp_norm(X3,Y3,3)

# L4 norm
X4,Y4 = np.meshgrid(v1, v2)
Z4 = lp_norm(X4,Y4,4)

# Max norm
X_max,Y_max = np.meshgrid(v1, v2)
Z_max = lp_norm(X_max,Y_max,100)

# L(0.5) norm
X_half,Y_half = np.meshgrid(v1, v2)
Z_half = lp_norm(X_half,Y_half,0.5)
```

```python
# Plotting
CS1 = plt.contour(X1, Y1, Z1, [1], colors = 'r', linestyles = 'solid')
CS2 = plt.contour(X2, Y2, Z2, [1], colors = 'g', linestyles = 'solid')
CS3 = plt.contour(X3, Y3, Z3, [1], colors = 'c', linestyles = 'solid')
CS4 = plt.contour(X4, Y4, Z4, [1], colors = 'y', linestyles = 'solid')
CS_max = plt.contour(X_max, Y_max, Z_max, [1], colors = 'b', linestyles = 'solid')
CS_half = plt.contour(X_half, Y_half, Z_half, [1], colors = 'k', linestyles = 'solid')
plt.axes().set_aspect('equal') # Scale the plot size to get same aspect ratio
plt.axis([-1.2, 1.2, -1.2, 1.2]) # Change axis limits
plt.clabel(CS1, inline=1, fontsize=10)
plt.clabel(CS2, inline=1, fontsize=10)
plt.clabel(CS3, inline=1, fontsize=10)
plt.clabel(CS4, inline=1, fontsize=10)
plt.clabel(CS_max, inline=1, fontsize=10)
plt.title('Lp norm plots')
plt.grid(linewidth=0.5)
plt.show()


print('Legend: \n L1 norm -> red \n L2 norm -> green \n L3 norm -> cyan \n L4 norm -> yellow \n Max norm -> blue \n L(0.5) norm -> black')
```
